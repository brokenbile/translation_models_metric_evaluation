```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import pandas as pd
from tqdm import tqdm

tokenizer = AutoTokenizer.from_pretrained("facebook/nllb-200-distilled-600M", src_lang="fra_Latn", tgt_lang="en_Latn")
model = AutoModelForSeq2SeqLM.from_pretrained("facebook/nllb-200-distilled-600M")

from transformers import pipeline

tqdm.pandas()

text = pd.read_csv("en-fr-test.csv")
text = text[text['fr'].str.split().str.len().lt(300)] # Drop sentences with more than 300 words
print(text)

translator = pipeline("translation", model="Helsinki-NLP/opus-mt-fr-en", src_lang="fra_Latn", tgt_lang="en_Latn")
text["nllb"] = text["fr"].progress_apply(translator)

print(text)

```

```python
text["nllb"] = text["nllb"].progress_apply(lambda x: x[0]["translation_text"])

text.to_csv("nllb-translation-out.csv", index=False)
```
